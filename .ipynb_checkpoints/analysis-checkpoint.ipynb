```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_pickle('college_data.pkl')
```

```python
df
```

```python
df_grouped = df.groupby('school')
df_grouped = df_grouped.agg({
    'gpa': 'mean',          # Average GPA
    'act': 'mean',          # Average ACT
    'sat': 'mean',          # Average SAT
    'accepted': "sum",  # Sum of accepted (True converted to 1, False to 0)
    'appType': 'count'      # Count of applications
})
df_grouped = df_grouped.rename(columns={'appType': 'applicationCount'})
df_grouped['acceptanceRate'] = df_grouped['accepted'] / df_grouped['applicationCount']
```

```python
df_grouped = df_grouped.sort_values('acceptanceRate', ascending=True)
df_grouped[df_grouped['applicationCount'] > 10] # Top 25 schools with more than 10 applications
```

```python
# Joining and filtering for accepted students
# Specify suffixes to differentiate columns from the grouped data and original data
top_schools = df_grouped[df_grouped['applicationCount'] > 10].head(25).index.tolist()
df_accepted = df[(df['school'].isin(top_schools)) & (df['accepted'])]
df_accepted = df_accepted[df_accepted['sat'] != 0]
plt.figure(figsize=(10, 6))

# Create scatter plot (scattegram)
plt.scatter(df_accepted['sat'], df_accepted['gpa'], alpha=0.6, label='Data Points', color='blue')

# Calculating the line of best fit
# Fit the regression line using np.polyfit for a 1st degree polynomial (linear regression)
fit = np.polyfit(df_accepted['sat'], df_accepted['gpa'], deg=1)
# Use the coefficients obtained from the polyfit to generate y-values based on x-values
p = np.poly1d(fit)

# Plotting the line of best fit
plt.plot(df_accepted['sat'], p(df_accepted['sat']), "r--", label='Line of Best Fit')

# Enhancements to the plot
plt.title('SAT vs GPA for Accepted Students in Top 25 Schools')
plt.xlabel('SAT Score')
plt.ylabel('GPA')
plt.grid(True)
plt.legend()  # Display legend to clarify plot elements
plt.show()
```

```python
df_denied = df[(df['school'].isin(top_schools)) & (~df['accepted'])]
df_denied = df_denied[df_denied['sat'] != 0]

plt.figure(figsize=(10, 6))

# Create scatter plot (scattegram)
plt.scatter(df_denied['sat'], df_denied['gpa'], alpha=0.6, label='Data Points', color='blue')

# Calculating the line of best fit
# Fit the regression line using np.polyfit for a 1st degree polynomial (linear regression)
fit = np.polyfit(df_denied['sat'], df_denied['gpa'], deg=1)
# Use the coefficients obtained from the polyfit to generate y-values based on x-values
p = np.poly1d(fit)

# Plotting the line of best fit
plt.plot(df_denied['sat'], p(df_denied['sat']), "r--", label='Line of Best Fit')

# Enhancements to the plot
plt.title('SAT vs GPA for Denied Students in Top 25 Schools')
plt.xlabel('SAT Score')
plt.ylabel('GPA')
plt.grid(True)
plt.legend()  # Display legend to clarify plot elements
plt.show()

```

```python

# Filtering accepted and denied students


plt.figure(figsize=(10, 6))

# Scatter plot for denied students
plt.scatter(df_denied['sat'], df_denied['gpa'], color='red', label='Denied', alpha=0.6)

# Scatter plot for accepted students
plt.scatter(df_accepted['sat'], df_accepted['gpa'], color='green', label='Accepted', alpha=0.6)

# Enhancements to the plot
plt.title('SAT vs GPA for Top 25 Schools (Accepted & Denied)')
plt.xlabel('SAT Score')
plt.ylabel('GPA')
plt.grid(True)
plt.legend()  # Display legend to clarify plot elements
plt.show()
```
